```
def parse_XML(xml_file, df_cols): 
    """Parse the input XML file and store the result in a pandas 
    DataFrame with the given columns. 
    
    The first element of df_cols is supposed to be the identifier 
    variable, which is an attribute of each node element in the 
    XML data; other features will be parsed from the text content 
    of each sub-element. 
    """
    
    xtree = et.parse(xml_file)
    xroot = xtree.getroot(); print(xroot.text)
    rows = []
    
    for node in xroot: 
        res = []
        res.append(node.attrib.get(df_cols[0]))
        for el in df_cols[1:]: 
            if node is not None and node.find(el) is not None:
                res.append(node.find(el).text)
            else: 
                res.append(None)
        rows.append({df_cols[i]: res[i] 
                     for i, _ in enumerate(df_cols)})
    
    out_df = pd.DataFrame(rows, columns=df_cols)
        
    return out_df
```

```
import pandas as pd 
import xml.etree.ElementTree as et 

path = '/content/PUB_GenOutputbyFuelHourly.xml'
xtree = et.parse(path)
xroot = xtree.getroot()

for node in xroot: 
    s_date = node.attrib.get("Date")
    s_hour = node.find("Hour").text if node is not None else None
    s_nuclear = node.find("NUCLEAR").text if node is not None else None
    s_gas = node.find("GAS").text if node is not None else None
    s_hydro = node.find("HYDRO").text if node is not None else None
    s_wind = node.find("WIND").text if node is not None else None
    s_solar = node.find("SOLAR").text if node is not None else None
    s_biofuel = node.find("BIOFUEL").text if node is not None else None
    s_total = node.find("Total Output").text if node is not None else None
    
    rows.append({"Date": s_date, "Hour": s_hour, 
                 "NUCLEAR": s_nuclear, "GAS": s_gas, 'HYDRO': s_hydro, 
                 'WIND': s_wind, 'SOLAR': s_solar, 'BIOFUEL': s_biofuel, 
                 'Total Output': s_total})

out_df = pd.DataFrame(rows, columns = df_cols)
```
